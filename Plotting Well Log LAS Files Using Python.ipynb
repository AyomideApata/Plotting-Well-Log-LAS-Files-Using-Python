
Creating a Pandas Dataframe

```python
def volshale(gamma_ray, gamma_ray_max, gamma_ray_min):
    vshale = (gamma_ray - gamma_ray_min) / (gamma_ray_max - gamma_ray_min)
    return round(vshale, 4)

def density_porosity(input_density, matrix_density, fluid_density):
    denpor = (matrix_density - input_density) / (matrix_density - fluid_density)
    return round(denpor, 4)

def sw_archie(porosity, rt, rw, archieA, archieM, archieN):
    sw = ((archieA / (porosity ** archieM)) * (rw/rt))**(1/archieN)
    return sw
```

```python
well.reset_index (inplace = True)
well.rename (columns = {'DEPT':'DEPTH'}, inplace=True)
well
```

```python
def sw_archie(porosity, HDRS, RHOB, archieA, archieM, archieN):
    sw = ((archieA / (porosity ** archieM)) * (HDRS/RHOB))**(1/archieN)
    return sw
```

```python
well['SW'] = sw_archie(well['DPHI'], well['HDRS'], 0.622, 1, 2, 2)
well['SW_LIM'] = well['SW'].mask(well['SW']>1, 1)
well['Coal'] = np.where(well['RHOB']<=2.4, 0 ,1)
well
```



Plotting Log Data of Gamma Ray, Caliper curve, Resistivity Curve (Deep & Medium), Neutron and Density porosity curves,  Archie's water saturation and Bulk Density. 
Where each cureve tracks has there properties and features, and Coal curve is considered anything with a bulk density less than or equal to 2.4 g/cc bulk density.

```python
fig, ax = plt.subplots(figsize=(20,20)) 

## Sets up the figure space
ax1 = plt.subplot2grid((1,5), (0,0), rowspan=1, colspan = 1)
ax2 = ax1.twiny()
ax3 = plt.subplot2grid((1,5), (0,1), rowspan=1, colspan = 1)
ax4 = ax3.twiny()
ax5 = plt.subplot2grid((1,5), (0,2), rowspan=1, colspan = 1)
ax6 = ax5.twiny()
ax7 = plt.subplot2grid((1,5), (0,3), rowspan=1, colspan = 1)
ax8 = plt.subplot2grid((1,5), (0,4), rowspan=1, colspan = 1)
ax9 = ax8.twiny()

# Gamma Ray track
ax1.plot("GR", "DEPTH", data = well, color = "green")
ax1.set_xlabel("Gamma")
ax1.xaxis.label.set_color("green")
ax1.set_xlim(0, 150)
ax1.set_ylabel("Depth (m)")
ax1.tick_params(axis='x', colors="green")
ax1.spines["top"].set_edgecolor("green")
ax1.title.set_color('green')
ax1.set_xticks([0,25,50,75,100,125,150])

# Caliper curve
ax2.plot("CALI", "DEPTH", data = well, color = "red")
ax2.set_xlabel("Caliper")
ax2.set_xlim(0, 15)          
ax2.xaxis.label.set_color("red")
ax2.tick_params(axis='x', colors="red")
ax2.spines["top"].set_position(("axes", 1.08))
# ax2.spines["top"].set_edgecolor("red")
ax2.spines["top"].set_visible(True)
ax2.title.set_color('green')
ax2.set_xticks([0, 5, 10, 15])


## Resitivity 1
ax3.plot("HDRS", "DEPTH", data = well, color = "red")
ax3.set_xlabel("Deep Resitivity")
ax3.set_xlim(0.2, 2000)
ax3.xaxis.label.set_color("red")
ax3.tick_params(axis='x', colors="red")
ax3.spines["top"].set_edgecolor("red")
ax3.set_xticks([0.1,1,10,100,1000])
ax3.semilogx()
 
## Resitivity 2
ax4.plot("HMRS", "DEPTH", data = well, color = "purple")
ax4.set_xlabel("Medi Resitivity")
ax4.set_xlim(0.2 ,2000)
ax4.xaxis.label.set_color("purple")
ax4.tick_params(axis='x', colors="purple")
ax4.spines["top"].set_position(("axes", 1.08))
ax4.spines["top"].set_edgecolor("purple")
ax4.spines["top"].set_visible(True)
ax4.set_xticks([0.1,1,10,100,1000])
ax4.semilogx()
   
## Neutron track placed ontop of density track
ax5.plot("NPHI", "DEPTH", data = well, color = "blue")
ax5.set_xlabel("Neutron")
ax5.set_xlim(0.3, -0.1)
ax5.xaxis.label.set_color("blue")
ax5.tick_params(axis='x', colors="blue")
ax5.spines["top"].set_edgecolor("blue")
ax5.set_xticks([-0.1, 0, 0.1, 0.2, 0.3])
                      
ax6.plot("DPHI", "DEPTH", data = well, color = "red")
ax6.set_xlabel('Density')
ax6.xaxis.label.set_color("red")
ax6.set_xlim(0.5, 0)
ax6.tick_params(axis='x', colors="red")
ax6.spines["top"].set_position(("axes", 1.08))
ax6.spines["top"].set_visible(True)
ax6.spines["top"].set_edgecolor("red")
ax6.set_xticks([0,  0.25, 0.5])

## Sw track
ax7.plot("SW_LIM", "DEPTH", data = well, color = "black")
ax7.set_xlabel("SW - Archie")
ax7.set_xlim(0, 1)
ax7.xaxis.label.set_color("black")
ax7.tick_params(axis='x', colors="black")
ax7.spines["top"].set_edgecolor("black")
ax7.set_xticks([0, 0.5, 1])
               
## Bulk Density
ax8.plot("RHOB", "DEPTH", data = well, color = "black")
ax8.set_xlabel("Bulk")
ax8.set_xlim(1.95, 2.95)
ax8.xaxis.label.set_color("black")
ax8.tick_params(axis='x', colors="black")
ax8.spines["top"].set_edgecolor("black")
ax8.set_xticks([1.95, 2.45, 2.95])

## Coal
ax9.plot("Coal", "DEPTH", data = well, color = "blue")
ax9.set_xlabel("Coal")
ax9.set_xlim(0, 1)
ax9.xaxis.label.set_color("blue")
ax9.tick_params(axis='x', colors="blue")
ax9.spines["top"].set_position(("axes", 1.08))
ax9.spines["top"].set_visible(True)
ax9.spines["top"].set_edgecolor("blue")
ax9.set_xticks([0, 0.5, 1])


for ax in [ax1, ax3, ax5, ax7, ax8]:
    ax.set_ylim(11000, 0)
    ax.grid(which='major', color='lightgrey', linestyle='-')
    ax.xaxis.set_ticks_position("top")
    ax.xaxis.set_label_position("top")
    ax.spines["top"].set_position(("axes", 1.02))
    
    
plt.tight_layout()
```
